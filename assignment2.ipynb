```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "NLP is fun.",
    "NLP is powerful.",
    "Machine learning is fun."
]

vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(documents)

vocabulary = vectorizer.get_feature_names_out()

print("Vocabulary:")
print(vocabulary)

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())
```

```python
import pandas as pd

# Use 'r' before the string to handle the backslashes automatically
file_path = r"C:\Users\Admin\Desktop\nlp\labelled sentences(1).xlsx"

# Load the Excel file
df = pd.read_excel(file_path)

# Focus on the first column (index 0)
first_column = df.iloc[:, 0]

# Display the first few rows to confirm it loaded correctly
print(first_column.head())
```

```python
# 1. Select the first column and give it a clear name
text_data = df.iloc[:, 0].astype(str)

# 2. Check for and handle missing values
if text_data.isnull().sum() > 0:
    text_data = text_data.fillna("missing")
    print(f"Filled {text_data.isnull().sum()} missing rows.")
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary resources
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # 3. Tokenize (split into words)
    tokens = word_tokenize(text)
    # 4. Remove stopwords
    filtered_tokens = [w for w in tokens if w not in stop_words]
    
    return " ".join(filtered_tokens)

# Apply the function to your column
df['cleaned_text'] = text_data.apply(preprocess_text)

# Preview the result
print(df[['cleaned_text']].head())
```

```python
from collections import Counter

# 1. Combine all cleaned sentences into one large list of words
# We split each sentence into words and flatten the list
all_words = [word for sentence in df['cleaned_text'] for word in str(sentence).split()]

# 2. Count frequencies using Counter
vocab_counts = Counter(all_words)

# 3. Calculate Vocabulary Size
vocab_size = len(vocab_counts)

# 4. Get the Top 20 most frequent words
most_common = vocab_counts.most_common(20)

print(f"Total Unique Words (Vocabulary Size): {vocab_size}")
print("-" * 30)
print("Top 20 Most Frequent Words:")
for word, count in most_common:
    print(f"{word}: {count}")

# Optional: Convert to a DataFrame for better visualization
import pandas as pd
vocab_df = pd.DataFrame(most_common, columns=['Word', 'Frequency'])
```

```python
# 1. Convert the cleaned sentences into actual lists of words
# We split the string by whitespace to create a list for each row
df['word_list'] = df['cleaned_text'].apply(lambda x: x.split())

# 2. Select only this new column to keep the CSV clean
# If you want to keep the original text too, use df[['cleaned_text', 'word_list']]
output_df = df[['word_list']]

# 3. Save to a CSV file
# index=False prevents pandas from adding an extra column of row numbers
output_df.to_csv('processed_vocabulary_lists.xlsx.csv', index=False)

# Preview the first few rows of the new format
print("Preview of the word list format:")
print(output_df.head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd

# We assume 'df' is your DataFrame and 'cleaned_text' is the column from the previous step.
# If your column is named differently, update 'cleaned_text' below.
corpus = df['cleaned_text'].astype(str)

# --- 1. Count Vectorizer (Bag of Words) ---
# This counts how many times each word appears in each row
count_vec = CountVectorizer()
count_data = count_vec.fit_transform(corpus)

# Convert to a DataFrame for easy viewing
# get_feature_names_out() retrieves the unique words (the vocabulary)
count_df = pd.DataFrame(count_data.toarray(), columns=count_vec.get_feature_names_out())

# --- 2. TF-IDF Vectorizer ---
# This calculates the "importance" of a word relative to the whole dataset
tfidf_vec = TfidfVectorizer()
tfidf_data = tfidf_vec.fit_transform(corpus)

# Convert to a DataFrame for easy viewing
tfidf_df = pd.DataFrame(tfidf_data.toarray(), columns=tfidf_vec.get_feature_names_out())

# --- 3. Output/Preview ---
print("Count Vectorizer Shape:", count_df.shape)
print("TF-IDF Vectorizer Shape:", tfidf_df.shape)

# Preview the first 5 rows and 10 columns of the TF-IDF matrix
print("\nPreview of TF-IDF Weights (First 5 rows):")
display(tfidf_df.head())
```

```python
# 1. View the vocabulary as a Dictionary (Word -> Index)
# This is useful for checking if a specific word exists in your model
print("Vocabulary Dictionary (Sample):")
# We use list(dict.items())[:10] just to preview the first 10 entries
print(list(count_vec.vocabulary_.items())[:10])

# 2. View the Vocabulary as a List (Feature Names)
# This is the 'ordered' list of words used as headers for your vectors
feature_names = count_vec.get_feature_names_out()
print("\nFirst 20 words in alphabetical order:")
print(feature_names[:20])

# 3. Check the total Vocabulary Size
print(f"\nTotal unique words in vocabulary: {len(feature_names)}")

# 4. Search for a specific word's index
search_word = "vaccine"
if search_word in count_vec.vocabulary_:
    idx = count_vec.vocabulary_[search_word]
    print(f"\nThe word '{search_word}' is at index: {idx}")
else:
    print(f"\nThe word '{search_word}' was not found in the vocabulary.")
```
